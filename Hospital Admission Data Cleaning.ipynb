Import Libraries and File

```python
import pandas as pd
import numpy as np
from datetime import datetime

patient_record = pd.read_csv('dirty_healthcare_data.csv')
```

Capitalise the First Letter of the Patient Names

```python
patient_record['name'] = patient_record['name'].str.title()
```

Dealing with Age

```python
less_than_zero = patient_record['age'] < 18

print(patient_record.loc[less_than_zero,'age'].head(3))

```

```python

q1 = patient_record['age'].quantile(0.25)
q3 = patient_record['age'].quantile(0.75)

iqr = q3 - q1
upper_limit = q3 + (1.5 * iqr)
lower_limit = q1 - (1.5 * iqr)


age_outlier_mask = patient_record['age'] < 18
print(patient_record.loc[age_outlier_mask,['name','age']].head(3))

print('FIX APPLIED')

patient_record.loc[age_outlier_mask,'age'] = round(abs(lower_limit),0)
print(patient_record.loc[age_outlier_mask,['name','age']].head(3))
```

Standardize Gender

```python
patient_record['gender'].unique()

gender_map = {
    'F': 'Female',
    'female': 'Female',
    'male': 'Male',
    'M': 'Male',
    'Other ': 'Other',
    'N/A': np.nan
}

patient_record['gender'] = patient_record['gender'].replace(gender_map)

patient_record['gender'].unique()
```

Check for Invalid Heights

```python
height_outlier_mask = patient_record['height_cm'] < 0
print(patient_record.loc[height_outlier_mask,['name','height_cm']].head(3))

print('FIX APPLIED')

patient_record.loc[height_outlier_mask,'height_cm'] = patient_record['height_cm'] * -1
print(patient_record.loc[height_outlier_mask,['name','height_cm']].head(3))
```

Check for Invalid Weights

```python
weight_outlier_mask = patient_record['weight_kg'] < 0
print(patient_record.loc[weight_outlier_mask,['name','weight_kg']].head(3))

print('FIX APPLIED')

patient_record.loc[weight_outlier_mask,'weight_kg'] = patient_record['weight_kg'] * -1
print(patient_record.loc[weight_outlier_mask,['name','weight_kg']].head(3))
```

Calculate BMI and Replace BMI

```python
height_to_metres = patient_record['height_cm'] / 100

height_to_metre_sq = height_to_metres ** 2

bmi2 = patient_record['weight_kg'] / height_to_metre_sq
```

```python
# Check if the calculated bmi is the same as the bmi in the dataset
patient_record['bmi'] == bmi2

#Replace the original bmi with the calculated bmi
patient_record['bmi'] = bmi2
```

Blood Pressure Check

```python
patient_record['blood_pressure'] = patient_record['blood_pressure'].str.replace('over','/').str.replace(' / ','/').str.replace('mmHg','')

bp_split = patient_record['blood_pressure'].str.split('/',expand=True)

patient_record['systolic_bp'] = pd.to_numeric(bp_split[0],errors='coerce').astype('Int64')
patient_record['diastolic_bp'] = pd.to_numeric(bp_split[1],errors='coerce').astype('Int64')

```

Diagnosis Check

```python
patient_record['diagnosis'].unique()

patient_record['diagnosis'] = patient_record['diagnosis'].fillna('Unknown')

patient_record['diagnosis'].unique()
```

Date Check

```python
# Date Format 
print(patient_record['admission_date'].dtype) 
print(patient_record['discharge_date'].dtype) 

# For admission_date

mix_format = patient_record["admission_date"].str.fullmatch(r"\d{2}-[A-Za-z]{3}-\d{4}")

patient_record.loc[mix_format, 'admission_date'] = (
    pd.to_datetime(
        patient_record.loc[mix_format, 'admission_date'],
        format='%d-%b-%Y',      # US format
        errors='coerce'
    )
    .dt.strftime('%Y-%m-%d')    # standard ISO format
)

# Boolean mask: dates like '12 Aug 2025'
mask_space = patient_record["admission_date"].str.fullmatch(r"\d{2} [A-Za-z]{3} \d{4}")

# Use the mask to convert only those rows
patient_record.loc[mask_space, "admission_date"] = (
    pd.to_datetime(
        patient_record.loc[mask_space, "admission_date"],
        format="%d %b %Y",
        errors="coerce",
    )
    .dt.strftime("%Y-%m-%d")   # -> '2025-08-12'
)


mask = patient_record['admission_date'].astype(str).str.contains('/', na=False)

patient_record.loc[mask, 'admission_date'] = (
    pd.to_datetime(
        patient_record.loc[mask, 'admission_date'],
        format='%d/%m/%Y',      # US format
        errors='coerce'
    )
    .dt.strftime('%Y-%m-%d')    # standard ISO format
)

# Discharge Date
mix_format = patient_record["discharge_date"].str.fullmatch(r"\d{2}-[A-Za-z]{3}-\d{4}")

patient_record.loc[mix_format, 'discharge_date'] = (
    pd.to_datetime(
        patient_record.loc[mix_format, 'discharge_date'],
        format='%d-%b-%Y',      # US format
        errors='coerce'
    )
    .dt.strftime('%Y-%m-%d')    # standard ISO format
)

# Boolean mask: dates like '12 Aug 2025'
mask_space = patient_record["discharge_date"].str.fullmatch(r"\d{2} [A-Za-z]{3} \d{4}")

# Use the mask to convert only those rows
patient_record.loc[mask_space, "discharge_date"] = (
    pd.to_datetime(
        patient_record.loc[mask_space, "discharge_date"],
        format="%d %b %Y",
        errors="coerce",
    )
    .dt.strftime("%Y-%m-%d")   # -> '2025-08-12'
)

# For discharge_date
mask = patient_record['discharge_date'].astype(str).str.contains('/', na=False)

patient_record.loc[mask, 'discharge_date'] = (
    pd.to_datetime(
        patient_record.loc[mask, 'discharge_date'],
        format='%d/%m/%Y',
        errors='coerce'
    )
    .dt.strftime('%Y-%m-%d')
)

print(patient_record['admission_date'].dtype) 
print(patient_record['discharge_date'].dtype) 
```

```python
# Check that admision date is not after the discharge date

invalid_period = patient_record['admission_date'] > patient_record['discharge_date']
health_loc = patient_record.loc[invalid_period,['name', 'admission_date','discharge_date']].head(3)

print(health_loc)
```

Data Categorization

```python
patient_record['smoker'].unique()

smoker_map = {
    'former': 'Ex Smoker',
    'EX-SMOKER': 'Ex Smoker',
    'Former': 'Ex Smoker',
    'Y': 'Yes',
    'yes': 'Yes',
    'no': 'No',
    'N': 'No',
    np.nan: 'Unknown'
}

patient_record['smoker']= patient_record['smoker'].replace(smoker_map)

patient_record['smoker'].unique()
```

```python
# Add a new Column for Active Smokers

patient_record['active_smoker'] = patient_record['smoker'].replace('Ex Smoker','No')

patient_record['active_smoker'].unique()

```

Standardise the Billing Amount - column should be float

```python
dirty_billing = (
    patient_record["billing_amount_usd"]
    .astype(str)
    .str.contains(r"\$|USD", na=False)
)

print(patient_record.loc[dirty_billing,['patient_id','billing_amount_usd']].head(3))

patient_record['billing_amount_usd'] = patient_record['billing_amount_usd'].astype(str).str.replace('[^\d.-]|USD ','',regex=True)
patient_record['billing_amount_usd'] =pd.to_numeric(patient_record['billing_amount_usd'])


print(patient_record.loc[dirty_billing,['patient_id','billing_amount_usd']].head(3))

#patient_record['billing_amount_usd']
```
